```python
import pandas as pd
import re
```

```python
df = pd.read_csv('job_offers.csv')
```

```python
df
```

```python
keywords = 'data|dados|analyst|analista|analytics|BI|business intelligence|ML|machine learning|AI|artificial intelligence|inteligência artificial|statistical'

clean_df = df[df['job_title'].str.contains(keywords, case=False, na=False)]
```

```python
clean_df
```

```python
def simplify(role):
    role = role.lower()
    analyst = re.compile('analyst|analytics|analista')
    engineer = re.compile('engineer|engineering|engenheiro|engenharia')
    bi = re.compile('BI|business intelligence|business')
    if bi.search(role):
        role = 'Analista BI'
    elif analyst.search(role):
        role = 'Analista de Dados'
    elif engineer.search(role):
        role = 'Engenheiro de Dados'
    else:
        role = 'Cientista de Dados'
    return role

clean_df['role'] = [simplify(role) for role in clean_df['job_title']]
```

```python
clean_df
```

```python
def education(desc, level):
    desc = desc.lower()
    bachelor = re.compile('bachelor|bs.c|graduação|superior completo|superior em|superior cursando|ensino superior|curso superior|formação superior|nível superior')
    masters = re.compile('master’s|ms.c|msc|mestrado')
    phd = re.compile('phd|ph.d|doutorado|pós graduação|pós-graduação')
    if level=='b':
        if bachelor.search(desc):
            return 1
        else:
            return 0
    if level=='m':
        if masters.search(desc):
            return 1
        else:
            return 0
    if level=='p':
        if phd.search(desc):
            return 1
        else:
            return 0

clean_df['bachelor'] = [education(desc, 'b') for desc in clean_df['job_desc']]
clean_df['masters'] = [education(desc, 'm') for desc in clean_df['job_desc']]
clean_df['phd'] = [education(desc, 'p') for desc in clean_df['job_desc']]
```

```python
def skills(desc, skill):
    desc = desc.lower()
    python = re.compile('python')
    sql = re.compile('sql')
    nosql = re.compile('nosql')
    java = re.compile('java')
    julia = re.compile('julia')
    matlab = re.compile('matlab')
    tensorflow = re.compile('tensorflow')
    tableau = re.compile('tableau')
    powerbi = re.compile('powerbi|power bi')
    qlikview = re.compile('qlikview|qlik view')
    excel = re.compile('ms excel|microsoft excel|microsoft office|ms office|pacote office')
    cloud = re.compile('aws|azure|gcp|ibm cloud')
    
    if skill=='python':
        if python.search(desc):
            return 1
        else:
            return 0
    if skill=='sql':
        if sql.search(desc):
            return 1
        else:
            return 0
    if skill=='nosql':
        if nosql.search(desc):
            return 1
        else:
            return 0
    if skill=='java':
        if java.search(desc):
            return 1
        else:
            return 0
    if skill=='julia':
        if julia.search(desc):
            return 1
        else:
            return 0
    if skill=='matlab':
        if matlab.search(desc):
            return 1
        else:
            return 0
    if skill=='tensorflow':
        if tensorflow.search(desc):
            return 1
        else:
            return 0
    if skill=='tableau':
        if tableau.search(desc):
            return 1
        else:
            return 0
    if skill=='powerbi':
        if powerbi.search(desc):
            return 1
        else:
            return 0
    if skill=='qlikview':
        if qlikview.search(desc):
            return 1
        else:
            return 0
    if skill=='excel':
        if excel.search(desc):
            return 1
        else:
            return 0
    if skill=='cloud':
        if cloud.search(desc):
            return 1
        else:
            return 0

clean_df['python'] = [skills(desc, 'python') for desc in clean_df['job_desc']]
clean_df['sql'] = [skills(desc, 'sql') for desc in clean_df['job_desc']]
clean_df['nosql'] = [skills(desc, 'nosql') for desc in clean_df['job_desc']]
clean_df['java'] = [skills(desc, 'java') for desc in clean_df['job_desc']]
clean_df['julia'] = [skills(desc, 'julia') for desc in clean_df['job_desc']]
clean_df['matlab'] = [skills(desc, 'matlab') for desc in clean_df['job_desc']]
clean_df['tensorflow'] = [skills(desc, 'tensorflow') for desc in clean_df['job_desc']]
clean_df['tableau'] = [skills(desc, 'tableau') for desc in clean_df['job_desc']]
clean_df['powerbi'] = [skills(desc, 'powerbi') for desc in clean_df['job_desc']]
clean_df['qlikview'] = [skills(desc, 'qlikview') for desc in clean_df['job_desc']]
clean_df['excel'] = [skills(desc, 'excel') for desc in clean_df['job_desc']]
clean_df['cloud'] = [skills(desc, 'cloud') for desc in clean_df['job_desc']]
```
